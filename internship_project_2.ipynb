```python
import yfinance as yf
import pandas as pd

stocks = ["AAPL", "MSFT", "GOOG"]

data = yf.download(
    stocks,
    start="2022-01-01",
    auto_adjust=True
)["Close"]

print(data.head(10))
```

```python
print(len(data))
```

```python
returns = data.pct_change().dropna()
print(returns)
```

```python
cov_matrix = returns.cov()
mean_returns = returns.mean()
print(cov_matrix)
print(mean_returns)
```

```python
import numpy as np

weights = np.array([0.4, 0.3, 0.3])
portfolio_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
portfolio_std = np.sqrt(portfolio_variance)
print(portfolio_std)
```

```python
rolling_vol = returns.rolling(window=30).std()
print(rolling_vol)
```

```python
correlation = returns.corr()
print(correlation)
```

```python
simulations = 10000
T = 252  # trading days
portfolio_simulations = np.zeros(simulations)

for i in range(simulations):
    random_returns = np.random.multivariate_normal(mean_returns, cov_matrix, T)
    portfolio_path = np.cumprod(np.dot(random_returns, weights) + 1)
    portfolio_simulations[i] = portfolio_path[-1]
```

```python
VaR_95 = np.percentile(portfolio_simulations, 5)
```
